```python
import pandas as pd
import glob

# Define the threshold
threshold = 0.05

# List of files to process
files = glob.glob('*.tsv')

# Initialize an empty list to store DataFrames
dataframes = []

# Loop through each file
for file in files:
    # Read the file into a DataFrame
    df = pd.read_csv(file, sep="\t", header=None,
                     names=["protein", "sequence", "seq_len", "analysis", "signature_acc", "signature_desc", "start", "stop", "score", "status", "date", "interpro_acc", "interpro_desc", "GO_annotations", "Pathways_annotations"])
    
    # Filter out rows where 'score' is '-'
    df = df[df['score'] != '-']
    
    # Convert 'score' to float
    df['score'] = df['score'].astype(float)
    
    # Filter rows based on the threshold
    df = df[df['score'] < threshold]
    
    # Select specific columns and drop duplicates
    df = df[['analysis', 'signature_acc', 'signature_desc', 'interpro_acc', 'interpro_desc']].drop_duplicates().reset_index(drop=True)
    
    # Add a column with the filename as the sample identifier
    df['sample_identifier'] = file
    
    # Append the DataFrame to the list
    dataframes.append(df)

# Concatenate all DataFrames into a single DataFrame
interpro_domains_full = pd.concat(dataframes, ignore_index=True)

# Display the result
#print(result_df)
interpro_domains_full.head()
```
